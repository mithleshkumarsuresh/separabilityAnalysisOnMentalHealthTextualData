```python
import pandas as pd
import numpy as np
import re
import nltk

from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')
import emoji
```

```python
dep = pd.read_csv('Suicide Sentiment Data\\Negative_Neutral_Depression\\Depression.csv')
```

```python
dep
```

```python
dep.describe(include='all')
```

```python
dep["Unnamed: 2"].value_counts()
```

```python
txt = (dep[dep["Unnamed: 2"] == 's']['TEXT']+dep[dep["Unnamed: 2"] == 's']['Unnamed: 2'])
txt.values
```

```python
dep["TEXT"] = dep["TEXT"].add(dep["Unnamed: 2"], fill_value='')
```

```python
dep[dep["TEXT"].isnull()]
```

```python
dep["Label"].value_counts() == len(dep)
```

Now, we can safely drop the duplicates as the labels are always 1 for all text in this dataframe and The null values of TEXT are not helpful in our textual analysis part too.

*Removal of Missing Values*

```python
dep.dropna(subset=["TEXT"],inplace=True)
```

```python
dep
```

*Text Cleaning*

```python
def cleaningText(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = re.sub(r"'", "", text)
    text = re.sub(r'\W+', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
dep["TEXT"] = dep["TEXT"].apply(cleaningText)
```

```python
dep
```

*Tokenization*

```python
def tokenizeText(text):
    return word_tokenize(text)
```

```python
list(word_tokenize(dep.iloc[0]["TEXT"]))
```

```python
dep["TEXT"] = dep["TEXT"].apply(tokenizeText)
```

```python
dep.iloc[1].values
```

The data analysis part will be taken part to check if we should remove the stop words and eventually carry on to the stemming and lemmatization

```python
dep.to_csv("depressionCleanedDataset.csv")
```
